```python
import pandas as pd

df = pd.read_csv("ucl_stats.csv", sep=";", skiprows=[0,1])

df.columns = [
    "Rank", "Squad", "MP", "W", "D", "L",
    "GF", "GA", "GD", "Pts", "xG", "xGA", "xGD", "xGD/90", "Notes"
]
```

```python
# On convertit bien en numérique
df["GF"] = pd.to_numeric(df["GF"], errors="coerce")
df["xG"] = pd.to_numeric(df["xG"], errors="coerce")

# Rapport efficacité
df["Efficiency_GF_xG"] = df["GF"] / df["xG"]

# Classement
efficiency = df.sort_values("Efficiency_GF_xG", ascending=False)[["Squad", "GF", "xG", "Efficiency_GF_xG"]]
print(efficiency.head(10))  # Top 10 équipes
```

```python
# On convertit bien en numérique
df["GF"] = pd.to_numeric(df["GF"], errors="coerce")
df["xG"] = pd.to_numeric(df["xG"], errors="coerce")

# Rapport efficacité
df["Efficiency_GF_xG"] = df["GF"] / df["xG"]

# Classement
efficiency = df.sort_values("Efficiency_GF_xG", ascending=False)[["Squad", "GF", "xG", "Efficiency_GF_xG"]]
print(efficiency.tail(10))  # Top 10 équipes
```

```python
# Efficacité offensive = Buts marqués / xG
df["Efficiency_GF_xG"] = df["GF"] / df["xG"]

# Top 10 équipes les + efficaces
top_eff = df.nlargest(10, "Efficiency_GF_xG")[["Squad", "GF", "xG", "Efficiency_GF_xG"]]

# Top 10 équipes les - efficaces
worst_eff = df.nsmallest(10, "Efficiency_GF_xG")[["Squad", "GF", "xG", "Efficiency_GF_xG"]]

print("Meilleures équipes (efficacité offensive) :")
print(top_eff)

print("\nMoins bonnes équipes (inefficacité offensive) :")
print(worst_eff)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x="Efficiency_GF_xG", y="Squad", data=top_eff, palette="viridis")
plt.title("Top 10 équipes les + efficaces (GF / xG)")
plt.show()
```

```python
sns.barplot(
    x="Efficiency_GF_xG",
    y="Squad",
    data=top_eff,
    hue="Squad",         # on met Squad comme hue
    dodge=False,         # pour éviter que ça décale les barres
    legend=False,        # on cache la légende
    palette="viridis"
)
```

```python
# Bottom 10 : équipes les moins efficaces devant le but
worst_eff = df.nsmallest(10, "Efficiency_GF_xG")[["Squad", "GF", "xG", "Efficiency_GF_xG"]]
print(worst_eff)

# Visualisation
plt.figure(figsize=(10,6))
sns.barplot(
    x="Efficiency_GF_xG",
    y="Squad",
    data=worst_eff,
    hue="Squad",
    dodge=False,
    legend=False,
    palette="magma"   # une autre palette pour bien différencier
)
plt.title("10 équipes les moins efficaces devant le but (GF / xG)")
plt.xlabel("Efficacité (GF / xG)")
plt.ylabel("Équipe")
plt.show()
```

```python
# Estimation des points attendus (xPts)
df["xPts"] = (df["xG"] / (df["xG"] + df["xGA"])) * (df["MP"] * 3)

# Surperf / Underperf = Points réels - Points attendus
df["Perf_gap"] = df["Pts"] - df["xPts"]

# Top 10 surperformers
top_perf = df.nlargest(10, "Perf_gap")[["Squad", "Pts", "xPts", "Perf_gap"]]
print("Top surperformers:")
print(top_perf)

# Top 10 underperformers
worst_perf = df.nsmallest(10, "Perf_gap")[["Squad", "Pts", "xPts", "Perf_gap"]]
print("\nTop underperformers:")
print(worst_perf)

# Visualisation
plt.figure(figsize=(12,6))
sns.barplot(
    x="Perf_gap",
    y="Squad",
    data=df.sort_values("Perf_gap", ascending=False),
    hue="Squad",
    dodge=False,
    legend=False,
    palette="coolwarm"
)
plt.axvline(0, color="black", linestyle="--")
plt.title("Surperformance vs Sous-performance (Pts - xPts)")
plt.xlabel("Écart entre points réels et attendus")
plt.ylabel("Équipe")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# Scatter plot xG vs xGA
sns.scatterplot(x="xGA", y="xG", data=df, s=100)

# Ajouter les labels (noms des équipes)
for i, row in df.iterrows():
    plt.text(row["xGA"]+0.05, row["xG"]+0.05, row["Squad"], fontsize=8)

# Axes inversés pour xGA (plus bas = mieux défensivement)
plt.gca().invert_xaxis()

# Titre & labels
plt.title("Profil des équipes en Ligue des Champions (xG vs xGA)", fontsize=14)
plt.xlabel("xGA (expected goals concédés)")
plt.ylabel("xG (expected goals créés)")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Exemple avec ton DataFrame df qui contient "xG" et "xGA"
x = df["xG"]
y = df["xGA"]

# Moyennes
x_mean = x.mean()
y_mean = y.mean()

fig, ax = plt.subplots(figsize=(8,6))

# Scatter plot
ax.scatter(x, y, s=5, color="black")

# Quadrants : colorier les zones
# Haut gauche
ax.axvspan(x.min(), x_mean, ymin=0, ymax=(y_mean-y.min())/(y.max()-y.min()), 
           facecolor="lightblue", alpha=0.3)
# Haut droite
ax.axvspan(x_mean, x.max(), ymin=0, ymax=(y_mean-y.min())/(y.max()-y.min()), 
           facecolor="lightgreen", alpha=0.3)
# Bas gauche
ax.axvspan(x.min(), x_mean, ymin=(y_mean-y.min())/(y.max()-y.min()), ymax=1, 
           facecolor="lightcoral", alpha=0.3)
# Bas droite
ax.axvspan(x_mean, x.max(), ymin=(y_mean-y.min())/(y.max()-y.min()), ymax=1, 
           facecolor="khaki", alpha=0.3)

# Quadrants
plt.text(0.02, 0.98, "Faible ", 
         transform=plt.gca().transAxes, color="red", ha='left', va='top')
plt.text(0.98, 0.98, "Créateurs d’occasions mais fragiles derrière", 
         transform=plt.gca().transAxes, color="yellow", ha='right', va='top')
plt.text(0.02, 0.02, "Ni inspirés, ni solides", 
         transform=plt.gca().transAxes, color="blue", ha='left', va='bottom')
plt.text(0.98, 0.02, "Machines complètes", 
         transform=plt.gca().transAxes, color="g", ha='right', va='bottom')

# Labels
for i, row in df.iterrows():
    ax.text(row["xG"]+0.1, row["xGA"]+0.1, row["Squad"], fontsize=9)

ax.set_xlabel("xG (Expected Goals)")
ax.set_ylabel("xGA (Expected Goals Against)")
ax.set_title("Scatter xG vs xGA avec quadrants colorés")

plt.show()
```
